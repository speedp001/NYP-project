````python
# Convert time strings to seconds using custom function
def convert_to_seconds(time_str):
    if pd.notna(time_str):
        parts = time_str.split(':')
        minutes = int(parts[0])
        seconds = float(parts[1])
        return minutes * 60 + seconds
    return np.nan

# Convert back to the original time format
def convert_to_time_format(seconds):
    if pd.notna(seconds):
        minutes, seconds = divmod(seconds, 60)
        return f"{int(minutes)}:{seconds:05.2f}"
    return np.nan
````

````python
# Convert the timings to seconds first

phelps_200m_df['200m Freestyle'] = phelps_200m_df['200m Freestyle'].apply(convert_to_seconds)
phelps_200m_df['200m Butterfly'] = phelps_200m_df['200m Butterfly'].apply(convert_to_seconds)
phelps_200m_df
````

````python
# Fill null values with mean values for '200m Freestyle' and '200m Butterfly'

phelps_200m_df['200m Freestyle'] = phelps_200m_df['200m Freestyle'].fillna(phelps_200m_df['200m Freestyle'].mean())
phelps_200m_df['200m Butterfly'] = phelps_200m_df['200m Butterfly'].fillna(phelps_200m_df['200m Butterfly'].mean())
````

````python
# Convert back to the original time format

phelps_200m_df['200m Freestyle'] = phelps_200m_df['200m Freestyle'].apply(convert_to_time_format)
phelps_200m_df['200m Butterfly'] = phelps_200m_df['200m Butterfly'].apply(convert_to_time_format)
phelps_200m_df
````
